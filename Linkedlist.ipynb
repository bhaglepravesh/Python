```python
class Node:
    
    def __init__(self,value):
        self.data=value
        self.next=None
        
class Linkedlist:
    
    def __init__(self):
        self.head=None
        
    
    def insertstart(self,value):
        new_node=Node(value)
        if self.head is None:
            self.head=new_node
        else:
            new_node.next=self.head
            self.head=new_node
    
    def reverses(self):
        l=[]
        temp=self.head
        while temp is not None:
            l.append(temp.data)
            temp=temp.next
        
        temps=self.head
        l.reverse()
        i=0
        while temps is not None:
            temps.data=l[i]
            i +=1
            temps=temps.next
        
        
    def insertend(self,value):
        new_node=Node(value)
        if self.head is None:
            self.head=new_node
        else:
            temp=self.head
            while temp.next is not None:
                temp=temp.next
            temp.next=new_node
            
    def insertafter(self,ele,value):
        new_node=Node(value)
        temp=self.head
        while temp.data!=ele:
            temp=temp.next
        new_node.next=temp.next
        temp.next=new_node
            
        
        
    def update(self,ele,value):
        temp=self.head
        while temp.data!=ele:
            temp=temp.next
        temp.data=value
        
    def deletefirst(self):
        
        self.head=self.head.next
        
    def deletelast(self):
        temp=self.head
        while temp.next.next is not None:
            temp=temp.next
        temp.next=None
   
    def deleteele(self,ele):
        if self.head.data==ele:
            self.head=self.head.next
        else:
            temp=self.head
            while temp.next.data!=ele:
                temp=temp.next
            temp.next=temp.next.next
        
    
    def display(self):
        temp=self.head
        size=0
        while(temp is not None):
            if(temp.next is not None):
                
                print(f"{temp.data}->",end="")
            else:
                print(temp.data)
            temp=temp.next
            size +=1
        print("size is :",size)
    
        
    
ll=Linkedlist()
ll.insertstart(1)
ll.insertstart(2)
ll.insertstart(3)
ll.insertstart(4)
ll.insertstart(5)
ll.display()
ll.insertend(6)
ll.insertend(7)
ll.insertend(8)
ll.display()
ll.insertafter(2,11)
ll.insertafter(8,9)
ll.display()
ll.update(1,21)
ll.update(5,25)
ll.update(9,19)
ll.display()
ll.deletefirst()
ll.deletefirst()
ll.display()
ll.deletelast()
ll.deletelast()
ll.display()
ll.deleteele(21)
ll.deleteele(3)
ll.deleteele(7)
ll.display()
ll.reverses()
ll.display()
```

```python
class Node:
    
    def __init__(self,value):
        self.data=value
        self.next=None
        self.prev=None
    
class DLL:
    
    def __init__(self):
        self.head=None
        self.tail=None
        
    def insertstart(self,value):
        
        new_node=Node(value)
        
        if self.head==None:
            self.head=new_node
            self.tail=new_node
        else:
            self.head.prev=new_node
            new_node.next=self.head
            self.head=new_node
            
    
    def insertend(self,value):
        new_node=Node(value)
        if self.head is None:
            self.head=new_node
            self.tail=new_node
        else:
            new_node.prev=self.tail
            self.tail.next=new_node
            self.tail=new_node
            
    def insertbefore(self,ele,value):
        
        new_node=Node(value)
        
        if self.head.data==ele:
            self.head.prev=new_node
            new_node.next=self.head
            self.head=new_node
        else:  
            temp=self.head
            while temp.next.data!=ele:
                temp=temp.next
            new_node.prev=temp
            new_node.next=temp.next
            temp.next=new_node
            temp.next.prev=new_node
            
                
        
        
    
    def display(self):
        temp=self.head
        count=0
        while temp is not None:
            if temp.next is None:
                print(temp.data)
            else:
                print(temp.data,"->",end="")
            temp=temp.next
            count +=1
        print("size of DLL :", count)
    
    
dll=DLL()

dll.insertstart(1)
dll.insertstart(2)
dll.insertstart(3)
dll.insertstart(4)
dll.insertstart(5)
dll.display()
print(dll.tail.data)

dll.insertend(6)
dll.insertend(7)
dll.insertend(8)
dll.display()
print(dll.head.data)
print(dll.tail.data)

dll.insertbefore(5,15)
dll.insertbefore(1,10)
dll.insertbefore(8,18)
dll.display()
```

```python
def max_xp_power(N, A, B, P):
    # DP array to store the maximum XP power
    dp = [0] * (N + 1)
    
    # Traverse the monsters from the end to the beginning
    for i in range(N - 1, -1, -1):
        max_xp = 0
        for j in range(A, B + 1):
            if i + j > N:
                break
            max_power = max(P[i:i+j])
            min_power = min(P[i:i+j])
            xp_gain = max_power - min_power
            if i + j < N:
                max_xp = max(max_xp, xp_gain + dp[i + j])
            else:
                max_xp = max(max_xp, xp_gain)
        dp[i] = max_xp
    
    return dp[0]

# Read input
N = int(input().strip())
A = int(input().strip())
B = int(input().strip())
P = [int(input().strip()) for _ in range(N)]

# Get the result
result = max_xp_power(N, A, B, P)
print(result)
```

```python
from collections import Counter

def has_anagram_subsequence(S, word):
    s_count = Counter(S)
    word_count = Counter(word)
    for char, count in word_count.items():
        if s_count[char] < count:
            return False
    return True

def solve(S,N, W):
    count = 0
    for word in W:
        if has_anagram_subsequence(S, word):
            count += 1
    return count

# Read input
S = input().strip()
N = int(input().strip())
W = [input().strip() for _ in range(N)]

# Get the result
result = solve(S,N, W)
print(result)
```
